## Part 1: Loading and Cleaning Data

### Loading Raw Data

```r
# Load libraries
library(dplyr)
library(ggplot2)
library(tidyverse)
```

```r
# Load datasets
setwd("../raw_data")
list.files()
```

```r
#load fish stock datasets
blackspot_seabream <- read.csv('blackspot seabream stock data.csv')
saithe <- read.csv('saithe stock data.csv')
```

```r
#clean blackspot_seabream dataset
blackspot_seabream_clean <- blackspot_seabream %>%
  rename(relative_exploitable_biomass = Relative.exploitable.biomass,
         x97.50 = X97.50.,
         x2.50 = X2.50.,
         landings_in_tonnes = Landings..tonnes.,
         relative_fishing_pressure = Relative.fishing.pressure,
         x9.501 = X97.50..1,
         x2.501 = X2.50..1)

head(blackspot_seabream_clean)
```

```r
# Plot seabream landings over the years
blackspot_seabream_clean %>% ggplot(aes(x = Year, y = landings_in_tonnes)) +
  geom_line() +
  geom_point() + 
  labs(title = "Blacksplot Seabream Landings Across Years",
       xlab = "Year",
       ylab = "Landings in Tonnes")
```

```r
#clean saithe dataset
saithe_clean <- saithe %>%
  rename(year = Year,
         landings_in_tonnes = Landings..tonnes.,
         discards_in_tonnes = Discards..tonnes.)

head(saithe_clean)
```

```r
# Plot seabream landings over the years
saithe_clean %>% ggplot(aes(x = year, y = landings_in_tonnes)) +
  geom_line() +
  geom_point()
```
